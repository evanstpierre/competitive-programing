1. What way should I **order** the set


#### Coin Problem

**Given:**
* A set of coints ${c_1,c_2,...,c_k}$ 
* A target $n$
**Find:**
* Form a sum of money $n$ using the coins

**Example:**
**Given:** coins, ${1,5,10,20,50,100,200}$ and $n=520$

**Solution:** $200 + 200 + 100 + 20$

**Greddy Strategy:** Select largest posibble coin until selected coins is the sum

**Key Assumption:** The solution set is non-empty


**Question: What sets of coins does this always work for?**

```python
# ASSUME: coins are sorted in ascending order
# ASSUME: solutions doe exist (n > 0)
coins = [1,5,10,20,50,100,200]
# target coins

n = 520
i = len(coins) - 1
# initalize empty solution set
sol = []

while(n != 0):
    if n >= coins[i]:
        sol.append(coins[i])
        n -= coins[i]
    else:
     i = max(i-1, 0)
print(sol)
```

#### Scheduling

**Given:** $n$ events with their start and end time

**Find:** Schedule that include as many events as possible

**Example:**
| Event | Starting Time | Ending Time |
|-------|----------------|--------------|
| A     | 1              | 3            |
| B     | 2              | 5            |
| C     | 3              | 9            |
| D     | 6              | 8            |

**Solution:** Maximumum number of events is $2$ (e.g. A, D) 

**Greddy Strategy:** Select the events of the algorithm that ends as soon as posible

**Key:** Sort Events by ending time (then apply greedy assumption)

```python
# Setting up envts  ([[event, start_time, end_time],...])

events = np.array([
    ["A", 1, 3],
    ["B", 2, 5],
    ["C", 3, 9],
    ["D", 6, 8]
])

# Sort rows by end time (column index 2)
events_sorted = np.array(sorted(events, key=lambda x: int(x[2])))

sol = []
t = 0   # current time

for i in range(len(events_sorted)):
    start = int(events_sorted[i, 1])
    end = int(events_sorted[i, 2])

    if start >= t:
        sol.append(events_sorted[i, 0])
        t = end

print(sol)
print(len(sol))
    
```

### Minimizing sums

**Given:** $n$ numbers, $a_1,a_2...,a_n$

**Find:** value $x$ such that it **minimizes** the sum

$$
\lvert a_1 - x \rvert^c + \lvert a_2 - x \rvert^c + \cdots + \lvert a_n - x \rvert^c 

$$

#### Case $c = 1$

Task: **Minimize** the sum,

$$
\lvert a_1 - x \rvert + \lvert a_2 - x \rvert + \cdots + \lvert a_n - x \rvert

$$


Let the following numbers be ${1,2,9,2,6}$. The optimal solution is *x=2*


$$
 \lvert 1 - 2 \rvert + \lvert 2 - 2 \rvert + \lvert 9 - 2 \rvert 
+ \lvert 2 - 2 \rvert + \lvert 6 - 2 \rvert = 12
$$

**Greedy Assumption:** the best x is the *median* of the numbers

**Proof Sketch:**

The median is an optimal choice, because 

**Case 1:** if $x$ is smaller than the median

the sum becomes smaller by increasing $x$,

**Case 2:** if $x$ is larger then the median, 

the sum becomes smaller by decreasing $x$. 


Hence, the optimal solution is that $x$ is
the median. 

**Note:** If n is even and there are two medians, both medians and all values
between them are optimal choices.

*source: 6.4 Minimizing Sums, Competative Programmers Handbook, 2018*

```python
# Solution:
numbers = np.array([1,2,9,2,6])

sol = int(np.median(numbers))

print(sol)
```

#### Case $c = 2$

Task: **Minimize** the sum,

$$
(a_1 - x)^2 + (a_2 - x)^2 + \cdots + (a_n - x)^2.
$$


Let the following numbers be ${1,2,9,2,6}$. The optimal solution is *x=2*

$$
(1 - 4)^2 + (2 - 4)^2 + (9 - 4)^2 + (2 - 4)^2 + (6 - 4)^2 = 46.
$$


**Greedy Assumption:** the best $x$ is the average of the numbers

**Proof Sketch:**

$$
n x^2 \;-\; 2x(a_1 + a_2 + \cdots + a_n) \;+\; (a_1^2 + a_2^2 + \cdots + a_n^2)
$$

Since we are miminimzing with respect to $x$. That means $a_1^2 + a_2^2 + \cdots + a_n^2$ is "constant"

so we have the following problem

$$
\min n x^2 \;-\; 2x(a_1 + a_2 + \cdots + a_n) 
$$

We can rewrite as follows, if we says $ s = a_1^2 + a_2^2 + \cdots + a_n^2$

$$
\min  n x^2  + 2x s
$$

The porabala $n x^2  + 2x s$  opens upwards and can be rewritten as $ x(nx-2s)$. 

Thererfore it has roots at $x=0$ and $x=2s/n$

the min values it the average of the roots roots $x = s/n$ (The average of  $a_1^2 + a_2^2 + \cdots + a_n^2$)

*source: 6.4 Minimizing Sums, Competative Programmers Handbook, 2018*

```python
# Solution:
numbers = np.array([1,2,9,2,6])

sol = round(np.average(numbers),1)

print(sol)
```

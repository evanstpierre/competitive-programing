### Minimizing sums

**Given:** $n$ numbers, $a_1,a_2...,a_n$

**Find:** value $x$ such that it **minimizes** the sum

$$
\lvert a_1 - x \rvert^c + \lvert a_2 - x \rvert^c + \cdots + \lvert a_n - x \rvert^c 

$$

#### Case $c = 1$

Task: **Minimize** the sum,

$$
\lvert a_1 - x \rvert + \lvert a_2 - x \rvert + \cdots + \lvert a_n - x \rvert

$$


Let the following numbers be ${1,2,9,2,6}$. The optimal solution is *x=2*


$$
 \lvert 1 - 2 \rvert + \lvert 2 - 2 \rvert + \lvert 9 - 2 \rvert 
+ \lvert 2 - 2 \rvert + \lvert 6 - 2 \rvert = 12
$$

**Greedy Assumption:** the best x is the *median* of the numbers

**Proof Sketch:**

The median is an optimal choice, because 

**Case 1:** if $x$ is smaller than the median

the sum becomes smaller by increasing $x$,

**Case 2:** if $x$ is larger then the median, 

the sum becomes smaller by decreasing $x$. 


Hence, the optimal solution is that $x$ is
the median. 

**Note:** If n is even and there are two medians, both medians and all values
between them are optimal choices.

*source: 6.4 Minimizing Sums, Competative Programmers Handbook, 2018*

#### Case $c = 2$

Task: **Minimize** the sum,

$$
(a_1 - x)^2 + (a_2 - x)^2 + \cdots + (a_n - x)^2.
$$


Let the following numbers be ${1,2,9,2,6}$. The optimal solution is *x=2*

$$
(1 - 4)^2 + (2 - 4)^2 + (9 - 4)^2 + (2 - 4)^2 + (6 - 4)^2 = 46.
$$


**Greedy Assumption:** the best $x$ is the average of the numbers

**Proof Sketch:**

$$
n x^2 \;-\; 2x(a_1 + a_2 + \cdots + a_n) \;+\; (a_1^2 + a_2^2 + \cdots + a_n^2)
$$

Since we are miminimzing with respect to $x$. That means $a_1^2 + a_2^2 + \cdots + a_n^2$ is "constant"

so we have the following problem

$$
\min n x^2 \;-\; 2x(a_1 + a_2 + \cdots + a_n) 
$$

We can rewrite as follows, if we says $ s = a_1^2 + a_2^2 + \cdots + a_n^2$

$$
\min  n x^2  + 2x s
$$

The porabala $n x^2  + 2x s$  opens upwards and can be rewritten as $ x(nx-2s)$. 

Thererfore it has roots at $x=0$ and $x=2s/n$

the min values it the average of the roots roots $x = s/n$ (The average of  $a_1^2 + a_2^2 + \cdots + a_n^2$)

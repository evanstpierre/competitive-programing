**Given:**
* A set of coints ${c_1,c_2,...,c_k}$

**Goal:** Solve problem efficiently for *any* coin set (General Case).

**Question:** Why does Dynamic Programming work?

*source: 7.1 Coin Problem, Competative Programmers Handbook, 2018*
